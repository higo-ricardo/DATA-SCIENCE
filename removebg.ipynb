```python
from rembg import remove
from PIL import Image
from google.colab import files
import os

# Permitir que o usuário carregue um arquivo do sistema
uploaded = files.upload()

# Obter o nome do arquivo carregado
input_filename = list(uploaded.keys())[0]
input_path = os.path.join("/content", input_filename)  # Caminho no Colab

# Nome do arquivo de saída (mantém o nome original e adiciona '_sem_fundo')
output_filename = f"sem_fundo_{os.path.splitext(input_filename)[0]}.png"
output_path = os.path.join("/content", output_filename)

# Abrir a imagem
image = Image.open(input_path)

# Remover o fundo
output = remove(image)

# Garantir que a imagem está no modo correto para suportar transparência
if output.mode != "RGBA":
    output = output.convert("RGBA")

# Salvar a imagem sem fundo corretamente
output.save(output_path, format="PNG")

# Permitir o download do arquivo processado
files.download(output_path)

print(f"Imagem sem fundo salva como {output_path} e disponível para download.")
```

2. O fundo será removido e um novo arquivo sem_fundo_<nome>.png será gerado.
3. O download do arquivo será iniciado automaticamente.

```python
import gradio as gr
from rembg import remove
from PIL import Image
import numpy as np
import io

# Função para remover o fundo e gerar o arquivo para download
def remove_background(input_image):
    # Converter a imagem do Gradio (numpy array) para PIL
    image = Image.fromarray(input_image)

    # Remover o fundo
    output_image = remove(image)

    # Salvar a imagem processada em um buffer
    output_buffer = io.BytesIO()
    output_image.save(output_buffer, format="PNG")  # PNG mantém a transparência
    output_buffer.seek(0)

    return np.array(output_image), output_buffer

# Criar a interface no Gradio
interface = gr.Interface(
    fn=remove_background,  # Função que processa a imagem
    inputs=gr.Image(type="numpy"),  # Entrada: Imagem do usuário
    outputs=[
        gr.Image(type="numpy", label="Imagem sem Fundo"),  # Exibir imagem processada
        gr.File(label="Baixar Imagem Sem Fundo")  # Botão de download
    ],
    title="Removedor de Fundo de Imagens",
    description="Faça upload de uma imagem e remova o fundo automaticamente! Baixe a imagem processada em PNG.",
    allow_flagging="never"
)

# Executar a interface
interface.launch()
```
