## QUESTÃO 3

**Jogo da Forca**

Você está criando um jogo da forca em Python, um jogo popular de adivinhação de palavras. Neste jogo, um jogador tenta adivinhar uma palavra, letra por letra, antes que o enforcado seja completamente desenhado.

**Instrução:**

Crie uma função chamada verificar_letra que aceite dois argumentos: uma palavra secreta e uma letra. A função deve verificar se a letra está presente na palavra secreta. Se a letra estiver na palavra secreta, a função deve retornar True, caso contrário, deve retornar False.

Apenas 1 letra.

```python
import random

GREEN = "\033[0;32m"
RED   = "\033[1;31m"
BLUE  = "\033[1;34m"
REVERSE = "\033[;7m"
RESET = "\033[0;0m"

# Função para escolher uma palavra aleatória de uma lista
def escolher_palavra():
 palavras = ["casa", "bola", "iPhone", "carne", "pseudofruto", "caneca", "universidade", "paralelepipedo", "computador", "Ricardo"]
 palavra_secreta = random.choice(palavras).upper()
 return palavra_secreta

# Função para mostrar a palavra secreta com mascara para letras não adivinhadas
def mostrar_palavra(palavra_secreta, letras_corretas):
  return "".join([letra if letra in letras_corretas else "__" for letra in palavra_secreta])

def jogo_da_forca():
  palavra_secreta = escolher_palavra().upper()
  letras_corretas = []
  letras_incorretas = []
  contador = len(palavra_secreta)

  print(REVERSE + "\t\t\t\tBEM-VINDO AO JOGO DA FORCA!")

  while True:

    mascara = mostrar_palavra(palavra_secreta, letras_corretas)
    print(f"\n\n\t\tVOCÊ {contador} CHANCES RESTANTES!")

    print(BLUE+"\nPALAVRA: " + mascara)
    print(RED+ "LETRAS INCORRETAS: " + ", ".join(letras_incorretas).upper())

    if mascara == palavra_secreta:
      print(GREEN +"\t\t\t\t\tPARABÉNS! VOCÊ GANHOU. \n\nA palavra secreta era: " + palavra_secreta)
      break

    if contador == 0:
      print(RED + "\t\t\t\t\tINFELIZMENTE, VOCÊ PERDEU ! \n\nA palavra secreta era: " + palavra_secreta)
      break

  # Solicitar uma letra do jogador
    letra = input("\nDIGITE UMA LETRA: ").upper()
    if len(letra) != 1 or not letra.isalpha():
      print("Por favor, digite uma única letra válida.")
      continue
    print("******************************************")
    if letra in palavra_secreta:
      letras_corretas.append(letra)
    else:
      letras_incorretas.append(letra)
      contador -=1

if __name__ == "__main__":
    jogo_da_forca()
```

# QUESTÃO 4

Você está desenvolvendo uma função que embaralha as letras de uma palavra para criar palavras embaralhadas.

**Instrução:**

Crie uma função chamada embaralhar_palavra que aceite uma palavra como argumento e retorne a mesma palavra com suas letras embaralhadas. Independentemente da entrada ser em maiúsculas ou minúsculas, a saída deve estar sempre em letras minúsculas.

A palavra deve ser: embaralhados

```python
import random
GREEN = "\033[0;32m"
RED   = "\033[1;31m"

def embaralhar_palavra(text):
  lista = list(text)
  random.shuffle(lista)
  novo_texto = ''.join(lista)
  return novo_texto.lower()

texto = "embaralhados"
palavra = embaralhar_palavra(texto)
print(RED + "\nPalavra inicial:", texto + GREEN + "\nPalavra final:", palavra)
```

Inicialize uma cartela de bingo com números de 1 a 20. A cartela pode ser uma lista de 2 listas com 5 números cada. Cada número deve ser único na cartela.

Utilize um loop para simular o sorteio de números de 1 a 20. Certifique-se de que os números sorteados sejam únicos.

Peça à pessoa que pressione "Enter" para sortear o próximo número a cada rodada.

sua cartela será:

```python
import random
GREEN = "\033[0;32m"
RED   = "\033[1;31m"
RESET = "\033[0;0m"


def sortear_bingo():
    numeros_sorteados = []

    # Inicialize uma cartela de bingo
    cartela = [[], [], []]


    for i in range(1, 13):
      x = random.randint(1,20)
      if len(cartela[0]) < 4:
        cartela[0].append(x)
      elif len(cartela[1])<4:
        cartela[1].append(x)
      else:
        cartela[2].append(x)


    # Função para exibir a cartela do jogador
    def exibir_cartela():
        print("Sua cartela de bingo:")
        for linha in cartela:
            print(linha)

    # Sorteio de números
    print("Pressione" + RED + " ENTER " + RESET + "para começar o sorteio do Bingo!\n")
    input()

    while len(numeros_sorteados) < 10:
        numero_sorteado = random.randint(1, 20)
        if numero_sorteado not in numeros_sorteados:
            numeros_sorteados.append(numero_sorteado)

            print(f"Número sorteado: {numero_sorteado}")
            exibir_cartela()
            input("Pressione Enter para sortear o próximo número.")

            for linha in cartela:
                if numero_sorteado in linha:
                    linha[linha.index(numero_sorteado)] = "X"

    print("Bingo!")

# Chame a função para começar o jogo de bingo
sortear_bingo()
```
